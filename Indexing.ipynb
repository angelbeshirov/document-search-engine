```python
def extractPdfFilesDocumentwise(files):
    loc = 1
    df = pd.DataFrame(columns = ("name", "content"))
    
    for file in files:
        print("Extracting document {}".format(loc))
        pdfFileObj = open(file, "rb")
        pdfReader = pycpdf.PDF(pdfFileObj.read())
        
        n_pages = len(pdfReader.pages)
        
        this_doc = ''
        for i in range(n_pages):
            pageObj = pdfReader.pages[i]
            this_text = pageObj.text.translate(pycpdf.unicode_translations)
            this_doc += this_text
        df.loc[loc] = file, this_doc
        loc = loc + 1
        
    return df
```

```python
def extractPdfFilesPagewise(files):
    loc = 1
    doc = 1
    df = pd.DataFrame(columns = ("name", "page", "content"))
    
    for file in files:
        print("Extracting document {}".format(doc))
        pdfFileObj = open(file, "rb")
        pdfReader = pycpdf.PDF(pdfFileObj.read())
        
        n_pages = len(pdfReader.pages)
        
        for i in range(n_pages):
            pageObj = pdfReader.pages[i]
            this_text = pageObj.text.translate(pycpdf.unicode_translations)
            df.loc[loc] = file, i + 1, this_text
            loc = loc + 1
        doc = doc + 1
        
    return df
```

```python
df = extractPdfFilesDocumentwise(files)
```
